The Entity-relationship diagram reveals the entities: pets, enclosures, food, customers, and transactions and their interactions with each other. The diagram portrays relationships in the form of diamonds. The relationships are feeding, purchasing, and being kept in. The last part of the diagram is the attributes. These details help contain important inofrmation under the entities. The purpose of the data set as a whole is to help monitor and administer particular events for the pet store. For every purchase of either a pet or food, the data is recorded. This includes the customers name and customer id (which goes along with the loyalty program). It also includes the transaction number, amount purchased, and the product number in other attributes. Thus, this entity-relationship diagram will help keep the company organized.
